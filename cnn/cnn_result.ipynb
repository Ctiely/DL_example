```python
import matplotlib.pyplot as plt
import random
from fashion_mnist_cnn import DataSet, CNNModel
```

```python
cnn = CNNModel((28, 28, 1), 10)
```

```python
test_data = DataSet('data/test.txt')
```

```python
datas = random.choices(test_data, k=16)
imgs = [data[0] for data in datas]
labels = [data[1] for data in datas]
preds = cnn.predict(imgs)
fig = plt.figure(figsize=(20, 20))
for i in range(16):
    ax = fig.add_subplot(4, 4, i + 1)
    ax.imshow(imgs[i]).set_cmap('gray')
    ax.get_xaxis().set_ticks([])
    ax.get_yaxis().set_ticks([])
    ax.set_title(f"True label {labels[i]}, Pred label {preds[i]}")
```
